```python
import random

clubs = []
diamonds = []
hearts = []
spades = []


class Card():
    
    def __init__(self,shape,value_str,value):
        
        self.shape = shape
        self.value_str = value_str
        self.value = value
        if value == 1:
            self.points = 11
        elif value > 1 and value < 11:
            self.points = value
        else:
            self.points = 10
            
    def __str__(self):
        
        return f'{self.value_str} of {self.shape}'



class Hand():
    
    def __init__(self):
        
        self.cards = []
        self.points = 0
        self.aces = 0
        
    def add_card(self,card):
        
        self.cards.append(card)
        self.points += card.points
        if card.value_str == 'Ace':
            self.aces += 1
            
    def adjust_for_ace(self):
        while self.points > 21 and self.aces:
            self.points -= 10
            self.aces -= 1
            
            
class Chips():
    
    def __init__(self):
        self.total = 100
        self.bet = 0
    
    def win_bet(self):
        self.total += self.bet
        
    def lose_bet(self):
        self.total -= self.bet
        
        
def take_bet(chips):
    print("Your total amount of chips is ",chips.total)
    while True:
        try:
            chips.bet = int(input('How many chips would you like to bet?  '))
        except:
            print('Please provide a number.')
        else:
            if chips.bet > chips.total:
                print("Bet can't exceed ",chips.total)
            else:
                break

        
def card_draw_from_deck():
    
    while True:
        
        shape = random.choice(['Clubs','Diamonds','Hearts','Spades'])
        value = random.randint(1, 13)

        if shape == clubs[0]:
            if clubs[value] != 'taken':

                karta = Card(clubs[0],clubs[value],value)
                clubs[value] = 'taken'
                return karta
            else:
                continue
                
        elif shape == diamonds[0]:
            if diamonds[value] != 'taken':

                karta = Card(diamonds[0],diamonds[value],value)
                diamonds[value] = 'taken'
                return karta
            else:
                continue
                
        elif shape == hearts[0]:
            if hearts[value] != 'taken':

                karta = Card(hearts[0],hearts[value],value)
                hearts[value] = 'taken'
                return karta
            else:
                continue
                
        elif shape == spades[0]:
            if spades[value] != 'taken':

                karta = Card(spades[0],spades[value],value)
                spades[value] = 'taken'
                return karta
            else:
                continue
                
                

def show_some(dealer,player):
    print("\nDealer's Hand:")
    print("  <card hidden>")
    print(f"  {dealer.cards[1]}")
    
    print("\nPlayer's Hand:", *player.cards, sep='\n  ')
    
    
def show_all(dealer,player):
    print("\nDealer's Hand:", *dealer.cards, sep='\n  ')
    print("\nPlayer's Hand:", *player.cards, sep='\n  ')
    
    
player_chips = Chips()

# ta pętla odpowiada za liczbę partyjek
while True:
    
    clubs = ['Clubs','Ace','Two','Three','Four','Five','Six','Seven','Eight','Nine','Ten','Jack','Queen','King']
    diamonds = ['Diamonds','Ace','Two','Three','Four','Five','Six','Seven','Eight','Nine','Ten','Jack','Queen','King']
    hearts = ['Hearts','Ace','Two','Three','Four','Five','Six','Seven','Eight','Nine','Ten','Jack','Queen','King']
    spades = ['Spades','Ace','Two','Three','Four','Five','Six','Seven','Eight','Nine','Ten','Jack','Queen','King']
    
    # the hand begins - partyjka się zaczyna
    print('\nWelcome to BlackJack!\n')
    
    # asking for the amount of chips
    
    take_bet(player_chips)
            
    
    # początkowe rozdanie kart---------------------------
    player_hand = Hand()
    player_hand.add_card(card_draw_from_deck())
    player_hand.add_card(card_draw_from_deck())
    
    dealer_hand = Hand()
    dealer_hand.add_card(card_draw_from_deck())
    dealer_hand.add_card(card_draw_from_deck())
    
    show_some(dealer_hand,player_hand)
    
    # I'm checking if there are 2 Aces in player's hand
    player_hand.adjust_for_ace()
    dealer_hand.adjust_for_ace()
    
    # PLAYER TURN @@@@@@@@@@@@@@@@@@@@@@@@@@@
    while True:
            
        # we're asking for the next move
        while True:
            move = input("\nWould you like to Hit or Stand? Enter 'h' or 's'  ")
            if move == 'h' or move == 's':
                break
                
        # player pressed 'h'
        if move == 'h':
            player_hand.add_card(card_draw_from_deck())
            player_hand.adjust_for_ace()
            # we display the cards
            show_some(dealer_hand,player_hand)
                
            #we check if player busts
            if player_hand.points > 21:
                print('\nPlayer busts!')
                player_chips.lose_bet()
                break
            
            
        else:
            print('Player stands. Dealer is playing.')
            # we display the cards
            show_all(dealer_hand,player_hand)
            break
        

    
    # DEALER TURN @@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@
    if player_hand.points <= 21:
    
        while dealer_hand.points < player_hand.points:
            # HIT
            dealer_hand.add_card(card_draw_from_deck())
            dealer_hand.adjust_for_ace()
            # we display the cards
            show_all(dealer_hand,player_hand)

        if dealer_hand.points > 21:
            print('\nDealer busts!')
            player_chips.win_bet()
            

        elif dealer_hand.points > player_hand.points:
            print('\nDealer wins!')
            player_chips.lose_bet()
            
        else:
            print("Dealer and Player tie! It's a push.")

    print("Player's winnings stand at ", player_chips.total)    
        
    while True:
        partia = input("\nWould you like to play another hand? Enter 'y' or 'n'  ")
        if partia == 'y' or partia == 'n':
            break
            
    if partia == 'n':
        print('Thanks for playing!')
        break
    
```
